## 1. Importar librerías necesarias

```python
# Importar librerías
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## 2. Cargar dataset
Elige una de estas opciones:
- Dataset de seaborn (`iris`, `tips`, `penguins`...)
- Dataset desde una URL (ejemplo: Titanic)
- Dataset propio en CSV


```python
# Ejemplo: dataset Iris
# Cambia este dataset por el que quieras trabajar
df = sns.load_dataset('iris')
df.head()
```

## 3. Exploración inicial del dataset

```python
# Explora las primeras filas
df.head()

# Información general del dataset
df.info()

# Estadísticas descriptivas
df.describe()
```

## 4. Calcular varianza

```python
# Calcula varianza de las variables numéricas
df.var(numeric_only=True)
```

## 5. Calcular covarianza

```python
# Calcula covarianza entre variables numéricas
df.cov(numeric_only=True)
```

## 6. Visualización de correlaciones

```python
# Mapa de calor de correlaciones
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Mapa de correlación entre variables")
plt.show()
```
